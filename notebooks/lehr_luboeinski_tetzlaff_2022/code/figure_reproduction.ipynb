## Content of this notebook

This notebook reproduces the following figures from **Lehr, Luboeinski & Tetzlaff, 2022**:

- Fig 2a,c,d (weights and rates)
- Fig 3h (firing rate distribution)
- Fig 5 (PCA)
- Fig 6 (spike time stability)
- Fig 7 (temporal sequences)
- Supp Fig 1 (weights and rates)
- Supp Fig 3 (PCA example)
- Supp Fig 4 (PCA different time bins)
- Supp Fig 5 (temporal sequences, core assembly only)
- Supp Fig 6 (temporal sequences, non-core neurons only)

The notebook makes use of a class written for this purpose called `stcAnalysisClass`, which can load and extract relevant data, perform analyses, and make the plots. Stastical tests reported in the paper can also be computed.

**Note.** The first time using this notebook, new directories will be created for the processed data and figures. The first time around, the extraction and analysis takes some time, e.g. for the regression part it can take hours. Then the data is stored in the `/processed_data` folder and figures in the `/figures` folder.

**IMPORTANT.** First get the raw data from LINK and put it in *lehr_luboeinski_tetzlaff_2022/raw_data/* to be able to run this notebook. If you also download the processed data and put it in *lehr_luboeinski_tetzlaff_2022/processed_data/* then most analysis will be skipped automatically and it will take the already analysed/processed data. This is of course faster if you do not want to reproduce everything.
